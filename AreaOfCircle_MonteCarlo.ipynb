# Area of a circle using Monte Carlo Simulation

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Equation of a unit circle centered at 1
# x^2+y^2=1

x_list = np.random.rand(1000);
y_list = np.sqrt(1-x_list*x_list)
```

```python
x_list_neg=-1*x_list
y_list_neg = -1*y_list

x_list = np.concatenate((x_list, x_list_neg), axis=0)
y_list = np.concatenate((y_list, y_list_neg), axis=0)
```

```python
x = np.linspace(-1.0, 1.0, 1000)
y = np.linspace(-1.0, 1.0, 1000)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 1
plt.contour(X,Y,F,[0])
plt.gca().set_aspect('equal')
```

```python
N_Trials = 100
number_of_hits=0;
t_x = [];
t_y = [];

for i in range(N_Trials):
    x = 2*(random.random())-1;
    y = 2*(random.random())-1;
    
    t_x.append(x);
    t_y.append(y);
    if x*x+y*y<=1:
        number_of_hits+=1;
```

```python
pi = number_of_hits/N_Trials;
print("Pi is", pi)
```

```python
t_x, t_y = np.meshgrid(t_x,t_y)
F = t_x**2 + t_y**2 - 1
plt.contour(t_x,t_y,F,[0])
plt.gca().set_aspect('equal')
```
